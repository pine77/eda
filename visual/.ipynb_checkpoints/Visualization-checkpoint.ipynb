### Line Chart
* plot() 
* 데이터의 추세..

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x,y)
plt.plot(x1, y1)

plt.savefig('test_vis.jpg')
```

```python
x1 = range(50)
y1 = [np.cos(value) for value in x1]

x2 = range(50)
y2 = [np.sin(value) for value in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x1, y1, color='red', linestyle='dotted')

plt.subplot(1,2,2)
plt.plot(x2, y2, color='blue', linestyle='dashed')
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+50 for value in x2]

plt.plot(x1, y1, c='#a52a2a', ls='-.', label='Line1')
plt.plot(x2, y2, c='#0000ff', ls = ':', label='Line2')

plt.text(50,60, 'Line 1')
plt.annotate('Line 2', xytext = (30, 110), xy = (50,100), arrowprops=dict(facecolor='yellow', shrink=0.1))

plt.legend(fancybox=True, shadow=True, loc='upper right')

plt.xlabel('X_axis')
plt.ylabel('Y_axis')
```

### Scatter Plot
* 데이터의 분포
* scatter

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.scatter(x,y)
```

```python
x = np.random.rand(200)
y = x + np.random.randn(200)/10

plt.scatter(x,y)
```

```python
data1 = np.random.rand(512,2)
data2 = np.random.rand(512,2)
```

```python
data1
```

```python
plt.scatter(data1[:, 0], data1[:, 1], c='b', marker='x')
plt.scatter(data2[:, 0], data2[:, 1], c='r', marker='o')
```

```python
n= 50
x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)
area = np.pi * (30*np.random.rand(n))**2

plt.scatter(x,y, c=colors, s=area, alpha=0.5)
plt.savefig('test_img.jpg')
```

### Bar Plot
* 데이터의 크기 비교
* bar()

```python
data = np.array([[5,25,50,20],[4,23,51,17],[6,22,52,19]])
x = np.arange(0,8,2)
```

```python
data
```

```python
x = np.array([0,2,4,6])
```

```python
y = data[0]
y
```

```python
plt.bar(x,y, color='b', width = 0.5)
```

```python
plt.bar(x,data[0], color='b', width = 0.5)
plt.bar(x,data[1], color='g', width = 0.5, bottom=np.sum(data[:1], axis=0))
plt.bar(x,data[2], color='r', width = 0.5, bottom=np.sum(data[:2], axis=0))
```

```python
plt.bar(x, data[0], color='b', width = 0.5)
plt.bar(x+0.5, data[1], color='g', width = 0.5)
plt.bar(x+1, data[2], color='r', width = 0.5)
```

##### Pandas Visualization

```python
# plt.figure()  - 도화지

# plt.plot() - 그래프 그리기

# plt.show() - 화면에 출력
```

```python
plt.figure(figsize=(10,5))

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(url)
drinks.head()
```

```python
drinks.columns = ['country','beer','spirit','wine','litres','continent']
drinks.head()
```

```python
df = drinks.copy()
```

```python
df.head()
```

```python
df = df.set_index('country')
df.head()
```

```python
# line chart
```

```python
df.beer.plot()
```

```python
df.sort_values('beer')['beer'].plot()
```

```python
# bar chart
```

```python
plt.figure(figsize=(15,3))
df.beer.plot(kind='bar')
```

```python
plt.figure(figsize=(15,3))
df.beer.sort_values(ascending=False)[:20].plot(kind='bar')
```

```python
# 대륙별 음료 평균소비량..
```

```python
df.groupby('continent').mean().plot(kind='bar')
```

```python
df.groupby('continent').mean().plot(kind='bar', stacked=True)
```

```python
# histogram
```

```python
df['beer'].plot(kind='hist', bins=30)
```

```python
# scatter
```

```python
df.plot(x= 'beer', y='wine', kind='scatter')
```

```python
pd.plotting.scatter_matrix(df)
plt.show()
```

```python
# boxplot
```

```python
df.plot(kind='box')
```

```python
# matshow()
```

```python
df.corr()
```

```python
plt.matshow(df.corr())
```
