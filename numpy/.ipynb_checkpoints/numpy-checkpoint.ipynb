```python
# Boolean Indexing

# array에 boolean array를  index로 하여 True에 해당하는 값만을 추출..
# array의 길이와 boolean array의 길이가 같아야 한다..
```

```python
ta = np.array([1,4,5,6,3,5,6])
```

```python
# ta에서 5 이상인 값만 출력하시오...
```

```python
ta >= 5
```

```python
ta[ta >= 5]
```

```python
# ta에서 짝수만 출력하시오..
```

```python
ta%2 == 0
```

```python
ta[ta%2 == 0]
```

```python
# 1 ~ 100 사이의 정수에서 20 보다 크고 30 보다 작은 3의 배수를 출력하시오..
# => boolean indexing을 사용..
```

```python
test_arr =  np.arange(1,101)
test_arr
```

```python
test_arr > 20
```

```python
test_arr < 30
```

```python
test_arr%3==0
```

```python
(test_arr > 20) & (test_arr < 30) & (test_arr%3==0)
```

```python
test_arr[(test_arr > 20) & (test_arr < 30) & (test_arr%3==0)]
```
